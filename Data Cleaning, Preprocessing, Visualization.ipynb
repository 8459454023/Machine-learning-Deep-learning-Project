```python
df["Memory"].value_counts()
```

```python
hdd = []
ssd = []

for i in df["Memory"]:
    if i[-3:] == "HDD":
        hdd.append(i)
        ssd.append("0")
    elif i[-3:] == "SSD":
        hdd.append("0")
        ssd.append(i)
    else:
        print("Problem somewhere")
```
